```python
global pos
pos = []
global pos3
pos3 = []
global pos2
pos2 = ""
global slip
slipz = []
```

```python
def slip(s):
    for i in range(1, len(s) + 1):
        slipz.append(i)
    for i in range(0, len(s)):
        for b in range(0, len(s)):
            if s[i] == s[b]:
                slipz[b] = slipz[i]
    return slipz
```

```python
def SBM(s, t):
    slip(s)
    step = 1
    start = int(len(s))-1
    end = int(len(t))
    w = int(len(s))-1
    lastchance = int(len(t[start:]))-1
    k=0
    if start == 0:
        start = start+1
    while lastchance > -1:
        for i in range(start,0,-1):
            if t[i] == s[w]:
                w-=1
                k+=1
                if k == int(len(s))-1:
                    print("Найдено")
                    lastchance = -5
                    break
                if k == int(len(s)) and int(len(s))-1 == 0:
                    print("Найдено")
                    lastchance = -5
                    break
            else:
                u = t[i]
                if s.find(u) != -1:
                    k = 0
                    step = slipz[s.index(u)]
                    start = start + step
                    lastchance = int(len(t[i:end-1]))
                    w = int(len(s))-1
                    if start > end-1:
                        for b in range(end-1,0,-1):
                            if t[b] == s[w]:
                                w -= 1
                                k += 1
                                if k == int(len(s)):
                                    print("Найдено")
                                    lastchance = -5
                                    break
                            else:
                                print("Не найдено")
                                lastchance = -5
                                break
                    break
                else:
                    k = 0
                    step = int(len(s))
                    start = start + step
                    lastchance = int(len(t[i:end-1]))
                    w = int(len(s)) - 1
                    if start > end-1:
                        for b in range(end-1,0,-1):
                            if t[b] == s[w]:
                                w -= 1
                                k += 1
                                if k == int(len(s)):
                                    print("Найдено")
                                    lastchance = -5
                                    break
                            else:
                                print("Не найдено")
                                lastchance = -5
                                break
                    if t[start-1] == s[w]:
                        for i in range(start-1, 0, -1):
                            if t[i] == s[w]:
                                w -= 1
                                k += 1
                                if k == int(len(s)) - 1:
                                    print("Найдено")
                                    lastchance = -5
                                    break
                    break
    choice2()
```

```python
def score():
    if int(len(pos)) == 0 and int(len(pos3))>0:
        pos2 = "Совпадений не найдено!"
        print(pos2)
        choice2()
    if int(len(pos)) > 0:
        pos2 = "Позиция(ии) слова - "
        pos2 += str(pos)
        print(pos2)
        choice2()
```

```python
def prefix(s):
    v = [0]*len(s)
    for i in range(1,len(s)):
        k = v[i-1]
        while k > 0 and s[k] != s[i]:
            k = v[k-1]
        if s[k] == s[i]:
            k = k + 1
        v[i] = k
    return v
```

```python
def kmp(s,t):
    global index
    index = -1
    global q
    q = s
    f = prefix(s)
    k = 0
    for i in range(len(t)):
        while k > 0 and s[k] != t[i]:
            k = f[k-1]
        if s[k] == t[i]:
            k = k + 1
        if k == len(s):
            index = i - len(s) + 1
            break
    if index != -1:
        pos.append(index)
    if index == -1:
        pos3.append(index)
    check(q,t)
    return index
```

```python
def check(s,t):
    global j
    j = ""
    len2 = t[int(index)+int(len(s)):]
    if (int(len(len2))) >= int(len(s)) and index != -1:
        for i in range(0,int(index)+int(len(s))):
            j = j + str("#")
        j += len2
        kmp(q,j)
        j = ""
    if (int(len(len2))) < int(len(s)):
        score()
```

```python
def start():
    global Stringz
    global SearchWord

    print("Введите строку: ")
    Stringz = input()
    print("Введите слово, которое хотите найти: ")
    SearchWord = input()
    choice(SearchWord,Stringz)
```

```python
def choice(sw,st):
    print("Выберите метод: 1.Кнута-Морриса-Пратта 2.Упрощенный Бойера-Мура")
    method = int(input())

    print("Важен ли регистр? Да/Нет")
    choice = input()
    if method == 1:
        if choice == "Да":
            kmp(sw,st)
        if choice == "Нет":
            sw = sw.lower()
            st = st.lower()
            kmp(sw,st)
    if method == 2:
        if choice == "Да":
            SBM(sw,st)
        if choice == "Нет":
            sw = sw.lower()
            st = st.lower()
            SBM(sw,st)
```

```python
def choice2():
    print("-------------------------")
    print("Хотите продолжить? Да/Нет")
    ch = input()
    if ch == "Да":
        pos.clear()
        pos3.clear()
        slipz.clear()
        start()
    if ch == "Нет":
        print("Ладно...")
```
