```python
def winner():
    string1 = "abc"
    string2 = "xya"

    string1 = ''.join(sorted(string1))
    string2 = ''.join(sorted(string2))
    x = 0
    y = 0
    for i in range(len(string1)):
        if string1[i] >= string2[i]:
            x += 1
        if string2[i] >= string1[i]:
            y += 1
        if x > y and x == 3:
            print("s2")
            break
        if x < y and y == 3:
            print("s1")
            break
        if x == y and i == len(string1)-1:
            print("")
            break
```

```python
winner()
```

```python
def polindrom():
    string3 = "abbabbbasooos"
    q = 0
    k = 0
    q2 = []
    f = 0
    for i in range(len(string3)-1):
        for b in range(i+1,len(string3)):
            if string3[i] == string3[b]:
                o = i
                p = b
                k = 1
                while o != b:
                    if string3[o] == string3[p]:
                        o += 1
                        p -= 1
                        k += 1
                    else:
                        break
                if k == 1+(b-i):
                    q2.append(string3[i:b+1])

    q = len(max(q2))

    for i in range(len(q2)):
        if len(q2[i]) == q:
            print(q2[i])
```

```python
polindrom()
```
