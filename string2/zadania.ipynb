```python
string1 = "abc"
string2 = "xya"

string1 = ''.join(sorted(string1))
string2 = ''.join(sorted(string2))
x = 0
y = 0
for i in range(len(string1)):
    if string1[i] >= string2[i]:
        x += 1
    if string2[i] >= string1[i]:
        y += 1
    if x > y and x == 3:
        print("s1")
        break
    if x < y and y == 3:
        print("s2")
        break
    if x == y and i == len(string1)-1:
        print("Равны")
        break
```
