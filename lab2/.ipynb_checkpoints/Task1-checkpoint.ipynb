```python
#Бинарый поиск
def binarySearch(alist, item):
        first_num = 0
        last_num = len(alist)-1
        found = False
        # перебор по среднему значению
        while last_num - first_num > 1:
                middle_num = math.floor((first_num + last_num) / 2)
                if alist[middle_num] < int(item):
                        first_num = middle_num
                if alist[middle_num] > int(item):
                        last_num = middle_num
                if alist[middle_num] == int(item) or alist[middle_num-1] == int(item) or alist[middle_num+1] == int(item):
                        found = True
                        break
        return found
```

```python
k = -2472970
q = time.time()
print(binarySearch(z,k))
print(time.time() - q)
```

```python
#Бинарное дерево
class Node:

    def __init__(self, data):

        self.left = None
        self.right = None
        self.data = data

# создание узлов(добавление)
    def insert(self, data):

        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
# сравнение значения с узлами
    def findval(self, serchnum):
        if serchnum < self.data:
            if self.left is None:
                return str(serchnum)+ " не найдено"
            return self.left.findval(serchnum)
        elif serchnum > self.data:
            if self.right is None:
                return str(serchnum)+ " не найдено"
            return self.right.findval(serchnum )
        else:
            print(str(self.data) + " было найдено")
```

```python
root = Node(12)

for i in range(len(x)-1):
    root.insert(x[i])
q = time.time()
print(root.findval(7))
print(root.findval(255))
print(time.time() - q)
```

```python
#Фибоначчи ------
def fibon():
    fib1 = 1
    fib2 = 1
    i = 0
    fib_sum = 0
    fib_mass = [1]
    while i != 10:  # генератор массива с числами фибоначчи
            fib_sum = fib1 + fib2
            fib1 = fib2
            fib2 = fib_sum
            fib_mass.append(fib_sum)
            i += 1
    return fib_mass
fibonx = fibon()
def fibonacci_search(array,n,index = -1):
    i = 0
    if (len(array) == 1):
        return index;
    while array[fibonx[i]] <= n:
        if fibonx[i+1] >= len(array):
            i = len(array)-1
            break
        i+=1
    index = i
    return fibonacci_search(array[i-1:i],n,index)
```

```python
q = time.time()
print (fibonacci_search(p,k))
print(time.time() - q)
```

```python
#Интерполяция
def interpolation(alist, item):
    first_num = 0
    last_num = len(alist)-1
    found = False
    k = 0
    # перебор по формуле
    while k != len(alist):
        middle_num = math.floor(first_num+(((int(item) - alist[first_num])*(last_num - first_num))/(alist[last_num] - alist[first_num])))
        if alist[middle_num] < int(item):
            first_num = middle_num + 1
        if alist[middle_num] > int(item):
            last_num = middle_num - 1
        if alist[middle_num] == int(item):
            found = True
            break
        k += 1
    return found        
```

```python
q = time.time()
print(interpolation(z,k))
print(time.time() - q)
```
